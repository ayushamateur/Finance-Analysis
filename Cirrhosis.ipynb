```python
df = pd.concat([df,sex, status, ahs], axis = 1)
```

```python
df.head()
```

```python
df.drop(['Status', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema'], axis = 1, inplace=True)
```

```python
df['Age'] = df['Age'].apply(lambda x : x//365)
```

```python
df.rename(columns={'Ascites_Y' : 'Ascites', 'Hepatomegaly_Y':'Hepatomegaly', 'Spiders_Y':'Spiders'}, inplace=True)
```

```python
X = df[['N_Days', 'Age', 'Bilirubin', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets',
        'Prothrombin', 'Stage','Ascites']]

y = df['D']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
predictions = log.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
log.coef_.reshape(12,1)
```
